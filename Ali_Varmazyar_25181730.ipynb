```python
# ==============================
# 1. Import Libraries
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
# Added Precision, Recall, and F1-Score
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Flatten

# Create directory to store saved models
if not os.path.exists('saved_models'):
    os.makedirs('saved_models')

# ==============================
# 2. Load Dataset
# ==============================
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/cicids2017_cleaned.csv")

data_sample = data.sample(n=10000, random_state=42)

X_full = data_sample.drop("Attack Type", axis=1)
y_full = LabelEncoder().fit_transform(data_sample["Attack Type"])

label_encoder = LabelEncoder()
y_full = label_encoder.fit_transform(data_sample["Attack Type"])
class_names = label_encoder.classes_

# Save the label encoder
joblib.dump(label_encoder, 'saved_models/label_encoder.pkl')

# ==============================
# 3. Settings
# ==============================
feature_counts = [45, 40, 35, 30]
epochs_list = [5, 10, 15]

results = []

# ==============================
# 4. Confusion Matrix Function
# ==============================
def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    cm_pct = cm.astype(float) / cm.sum(axis=1)[:, np.newaxis] * 100

    cm_df = pd.DataFrame(cm_pct, index=class_names, columns=class_names)
    print(f"\n{title} - Confusion Matrix (%)")
    display(cm_df.round(1))

    plt.figure(figsize=(8,6))
    sns.heatmap(cm_pct, annot=True, fmt=".1f", cmap="Blues")
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# ==============================
# 5. Training Loop
# ==============================
for k in feature_counts:
    print(f"\n===== {k} Features =====")

    selector = SelectKBest(score_func=f_classif, k=k)
    X_selected = selector.fit_transform(X_full, y_full)

    # Save the selector for this feature count
    joblib.dump(selector, f'saved_models/selector_k{k}.pkl')

    # --- FEATURE IMPORTANCE WITH NAMES ---
    all_scores = selector.scores_
    mask = selector.get_support()
    feature_names = X_full.columns[mask]
    feature_scores = all_scores[mask]

    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Score': feature_scores
    }).sort_values(by='Score', ascending=False)

    plt.figure(figsize=(12, 6))
    sns.barplot(data=importance_df, x='Feature', y='Score', palette='viridis', hue='Feature', legend=False)
    plt.title(f"Feature Importance (ANOVA) - Top {k}")
    plt.xticks(rotation=90, fontsize=9)
    plt.xlabel("Feature Name")
    plt.ylabel("ANOVA F-Score")
    plt.tight_layout()
    plt.show()

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_selected)

    # Save the scaler
    joblib.dump(scaler, f'saved_models/scaler_k{k}.pkl')

    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y_full, test_size=0.2, random_state=42
    )

    # -------- ML MODELS --------
    models = {
        "Decision Tree": DecisionTreeClassifier(),
        "Random Forest": RandomForestClassifier(),
        "KNN": KNeighborsClassifier(),
        "SVM": SVC()
    }

    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Calculate all required metrics
        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

        # Save ML model using joblib
        model_filename = f'saved_models/{name.replace(" ", "_")}_k{k}.pkl'
        joblib.dump(model, model_filename)

        plot_cm(y_test, y_pred, f"{name} ({k} features)")
        results.append({
            "Model": name, "Features": k, "Epochs": "-",
            "Accuracy": acc, "Precision": prec, "Recall": rec, "F1-Score": f1
        })

    # -------- DL MODELS --------
    X_train_dl = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_dl = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    for epoch in epochs_list:

        # LSTM
        lstm = Sequential([
            LSTM(64, input_shape=(X_train.shape[1],1)),
            Dense(len(np.unique(y_full)), activation='softmax')
        ])
        lstm.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
        lstm.fit(X_train_dl, y_train, epochs=epoch, verbose=0)

        y_pred = np.argmax(lstm.predict(X_test_dl), axis=1)

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

        # Save LSTM model
        lstm.save(f'saved_models/LSTM_k{k}_e{epoch}.h5')

        plot_cm(y_test, y_pred, f"LSTM ({k} features, {epoch} epochs)")
        results.append({
            "Model": "LSTM", "Features": k, "Epochs": epoch,
            "Accuracy": acc, "Precision": prec, "Recall": rec, "F1-Score": f1
        })

        # CNN
        cnn = Sequential([
            Conv1D(32, 3, activation='relu', input_shape=(X_train.shape[1],1)),
            MaxPooling1D(2),
            Flatten(),
            Dense(64, activation='relu'),
            Dense(len(np.unique(y_full)), activation='softmax')
        ])
        cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
        cnn.fit(X_train_dl, y_train, epochs=epoch, verbose=0)

        y_pred = np.argmax(cnn.predict(X_test_dl), axis=1)

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

        # Save CNN model
        cnn.save(f'saved_models/CNN_k{k}_e{epoch}.h5')

        plot_cm(y_test, y_pred, f"CNN ({k} features, {epoch} epochs)")
        results.append({
            "Model": "CNN", "Features": k, "Epochs": epoch,
            "Accuracy": acc, "Precision": prec, "Recall": rec, "F1-Score": f1
        })

# ==============================
# 6. Results Table
# ==============================
results_df = pd.DataFrame(results)
results_df["Accuracy (%)"] = results_df["Accuracy"] * 100
results_df["Precision (%)"] = results_df["Precision"] * 100
results_df["Recall (%)"] = results_df["Recall"] * 100
results_df["F1-Score (%)"] = results_df["F1-Score"] * 100

print("\nFinal Results Table:")
display(results_df)

# ==============================
# 7. ML Chart
# ==============================
plt.figure(figsize=(10,6))
ml_data = results_df[results_df["Epochs"] == "-"]

sns.lineplot(data=ml_data, x="Features", y="Accuracy (%)", hue="Model", marker="o")

for i in range(len(ml_data)):
    x = ml_data["Features"].iloc[i]
    y = ml_data["Accuracy (%)"].iloc[i]
    plt.text(x, y + 0.2, f"{y:.2f}%", ha='center', fontsize=8)

plt.title("ML Accuracy vs Number of Features")
plt.ylabel("Accuracy (%)")
plt.ylim(min(ml_data["Accuracy (%)"])-5, 105)
plt.grid(True)
plt.show()

# ==============================
# 8. DL Chart (Epochs)
# ==============================
plt.figure(figsize=(10,6))
dl_data = results_df[(results_df["Features"] == 30) &
                     (results_df["Model"].isin(["LSTM","CNN"]))]

sns.lineplot(data=dl_data, x="Epochs", y="Accuracy (%)", hue="Model", marker="o")

for i in range(len(dl_data)):
    x = dl_data["Epochs"].iloc[i]
    y = dl_data["Accuracy (%)"].iloc[i]
    plt.text(x, y + 0.2, f"{y:.2f}%", ha='center', fontsize=8)

plt.title("DL Accuracy vs Epochs (30 Features)")
plt.ylabel("Accuracy (%)")
plt.ylim(min(dl_data["Accuracy (%)"])-5, 105)
plt.grid(True)
plt.show()

# ==============================
# 9. DL Chart (Features vs Epochs)
# ==============================
for model_name in ["LSTM", "CNN"]:
    plt.figure(figsize=(10,6))

    model_data = results_df[results_df["Model"] == model_name]

    sns.lineplot(data=model_data, x="Epochs", y="Accuracy (%)",
                 hue="Features", marker="o", palette="viridis")

    for i in range(len(model_data)):
        x = model_data["Epochs"].iloc[i]
        y = model_data["Accuracy (%)"].iloc[i]
        plt.text(x, y + 0.4, f"{y:.1f}%", fontsize=8, ha='center', weight='bold')

    plt.title(f"{model_name} Accuracy vs Epochs (Different Features)")
    plt.ylabel("Accuracy (%)")
    plt.ylim(min(model_data["Accuracy (%)"])-5, 108)
    plt.grid(True)
    plt.show()

# ==============================
# 10. Final Comparison Chart (Multi-Metric)
# ==============================
best_results = results_df.loc[results_df.groupby("Model")["Accuracy"].idxmax()].copy()

best_results['Config'] = best_results.apply(
    lambda x: f"{x['Model']}\n(F:{x['Features']}, E:{x['Epochs']})", axis=1
)

melted_df = best_results.melt(id_vars="Config",
                             value_vars=["Accuracy (%)", "Precision (%)", "Recall (%)", "F1-Score (%)"],
                             var_name="Metric", value_name="Score")

plt.figure(figsize=(15,8))
ax = sns.barplot(data=melted_df, x="Config", y="Score", hue="Metric")

# Add text labels on bars
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.1f}%',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 12),
                    textcoords='offset points', fontsize=9, weight='bold')

plt.title("Final Performance Comparison: Best Model Configurations")
plt.ylabel("Score (%)")
plt.ylim(0, 125)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
